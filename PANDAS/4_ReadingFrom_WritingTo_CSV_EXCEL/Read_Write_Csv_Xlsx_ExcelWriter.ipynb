# Reading From EXCEL File

```python
df=pd.read_excel("stock_data.xlsx","Sheet1")
df
```

```python
def people_convert(cell):
    if cell=="n.a.":
        return "ASIM"
    return cell

def revenue_convert(cell):
    if cell==-1:
        return 0
    return cell

def eps_convert(cell):
    if cell=="not available" or cell=="n.a.":
        return None
    return cell
```

```python
df=pd.read_excel("stock_data.xlsx","Sheet1",converters={
    "people":people_convert,
    "revenue":revenue_convert,
    "eps":eps_convert,
    "price":eps_convert
})
df
```

# Writing To EXCEL File

```python
df.to_excel("new2.xlsx",sheet_name="STOCKS")
```

```python
df.to_excel("new2.xlsx",sheet_name="STOCKS",index=False)
```

# Merging two DataFrames into a single excel but different sheets

```python
Food=pd.DataFrame({
    "item":["shawarma","burger","pizza"],
    "price":[200,300,400]
})

Family=pd.DataFrame({
    "Members":["Asim","Afzal","Aqib"],
    "Age":[22,50,25]
})
```

```python
with pd.ExcelWriter("Family_Food.xlsx") as f:
    df.to_excel(f,sheet_name="Sht#1")
    df_final.to_excel(f,sheet_name="Sht#2")
    
```

```python
with pd.ExcelWriter("Family_Food.xlsx") as f:
    Food.to_excel(f,sheet_name="FOOD")
    Family.to_excel(f,sheet_name="FAMILY")
```
