# <font color="red">Dropna Method</font>

# <font color="red">Date_Range , DateTimeIndex and Reindexing Method</font>

**Now as you can see that some dates are missing and we want regularity in dates**

```python
df=pd.read_csv("weather_data.csv",parse_dates=["day"])
df=df.set_index("day")
df
```

```python
date_r=pd.date_range("1/1/2017","1/11/2017")
```

```python
date_index=pd.DatetimeIndex(date_r)
```

```python
new_df=df.reindex(date_index)
new_df
```
