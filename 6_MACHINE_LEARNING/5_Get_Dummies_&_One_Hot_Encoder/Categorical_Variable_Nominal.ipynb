```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df
```

# Get Dummies Method Of Python

```python
pd.get_dummies(df.town) # to make dummy varibles
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['town','west windsor'],axis=1)  # to avoid dummies variable trap -> 1 variable 2sre 2 se find kiya ja skta ho
final  # Agar ye trap rahe tw Linear Regression model sae se kam ni kta hai
```

```python
model = LinearRegression()
```

```python
x=final.drop(['price'],axis=1)
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2700,1,0]])
```

```python
model.predict([[3700,0,0]])
```

```python
model.score(x,y)
```

# One Hot Encoder Of Sklearn

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() # Alphabetical order mai label krta hai ye 
```

```python
dfle = df  # One Hot Encoder se phle Label encoder krna pre ga
dfle
```

```python
le.fit_transform(dfle.town)
```

```python
dfle.town=le.fit_transform(dfle.town)     # .values -> to convert into 2D array
dfle
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse=False) # if sparse True then neeche wali line mai .toarray() -> to convert into array
```

```python
dfohe=pd.DataFrame(ohe.fit_transform(dfle[['town']]),columns=["monroe township","west windsor","robinsville"])
dfohe
```

```python
df_ohe_final = dfohe.drop(["monroe township"],axis=1)
df_ohe_final
```

```python
dfle=dfle.drop('town',axis=1)
df_final=pd.concat([dfle,df_ohe_final],axis=1)
df_final
```

```python
x=df_final.drop(['price'],axis=1)
x
```

```python
y=df_final.price
y
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
