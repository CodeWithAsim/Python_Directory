# Histogram

Histogram is the bar chart with the frequency on the y axis

```python
import matplotlib.pyplot as plt
```

```python
Blood_Level=[45,65,67,87,45,98,89,112,123,134,144,150]  # In histogram we only need one dimension ... Y axis is the frequency automatically generated by the hist()
```

```python
plt.hist(Blood_Level)
```

By default histogram implements the 10 bins / buckets / ranges

```python
plt.hist(Blood_Level , bins=10)
```

```python
plt.hist(Blood_Level , bins=4)
```

we can also mention the ranges explicitly

```python
plt.hist(Blood_Level , bins=[60,80,100,120,150])
```

Now for gap between bars use rwidth property => relative width of bar with compare to the entire bin size / range 

```python
plt.hist(Blood_Level , bins=[60,80,100,120,150] , rwidth=0.95) # 95 percent of the entire
```

```python
plt.hist(Blood_Level , bins=[60,80,100,120,150] , rwidth=0.95 , color= "green") # or color="g"
```

```python
plt.hist(Blood_Level , bins=[60,80,100,120,150] , rwidth=0.95 , color= "green" , histtype="step")
```

Now if we want to draw 2 data set histogrms side by side then how , 

```python
Man_Blood_Level=[45,65,67,87,45,98,89,112,123,134,144,150]
Woman_Blood_Level=[35,55,57,77,35,88,78,102,113,124,134,140]
```

```python
plt.hist([Man_Blood_Level,Woman_Blood_Level] , bins=[60,80,100,120,150] , rwidth=0.95 , color= ["green","orange"] )
```

```python
plt.hist([Man_Blood_Level,Woman_Blood_Level] , bins=[60,80,100,120,150] , rwidth=0.95 , color= ["green","orange"] ,label=["Man","Woman"] )
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([Man_Blood_Level,Woman_Blood_Level] , bins=[60,80,100,120,150] , rwidth=0.95 , color= ["green","orange"] ,label=["Man","Woman"] )
plt.legend()
```

# Horizontal Orientation

```python
plt.ylabel("Sugar Level")
plt.xlabel("Number of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([Man_Blood_Level,Woman_Blood_Level] , bins=[60,80,100,120,150] , rwidth=0.95 , color= ["blue","orange"] ,label=["Man","Woman"] , orientation="horizontal" )
plt.legend()
```
