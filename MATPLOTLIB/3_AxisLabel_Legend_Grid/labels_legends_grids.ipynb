# <center><font color="red">Axis Labels _ Legends _ Grids </font></center>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day=[1,2,3,4,5,6,7]
max_t=[43,42,45,50,41,37,48]
min_t=[37,32,40,40,31,27,38]
avg_t=[40,35,42,45,36,33,43]
```

```python
plt.plot(day,max_t)
```

**Now you can plot the other three lines in the same canvas as follows :**

```python
plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,avg_t)
```

**Now we dont know what line is representing what , So we have to put legend ... For Legend we have to label the each line first then apply the legend property**

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(day,max_t,label="Max")
plt.plot(day,min_t,label="Min")
plt.plot(day,avg_t,label="Avg")
plt.legend()
```

**By default legend is set to best location ... We can set the location by using the loc property** 

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(day,max_t,label="Max")
plt.plot(day,min_t,label="Min")
plt.plot(day,avg_t,label="Avg")
plt.legend(loc="upper right")
```

**You can also set the font size in Legend**

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(day,max_t,label="Max")
plt.plot(day,min_t,label="Min")
plt.plot(day,avg_t,label="Avg")
plt.legend(loc="upper right" , fontsize="large")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(day,max_t,label="Max")
plt.plot(day,min_t,label="Min")
plt.plot(day,avg_t,label="Avg")
plt.legend(fontsize="small" , shadow=True) # Its All about fancy to play with legend 
```

**Now some people like grid to locate the data points**

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(day,max_t,label="Max")
plt.plot(day,min_t,label="Min")
plt.plot(day,avg_t,label="Avg")
plt.legend(fontsize="small" , shadow=True)
plt.grid()
```
