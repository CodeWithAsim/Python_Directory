# Format String

**Format strings are used to control the visual aspects of ypur 2D plot such as color , marker , markersize , linestyle etc**

```python
import matplotlib.pyplot as  plt
```

```python
x=[1,2,3,4,5,6,7]
y=[23,34,45,54,43,42,28]
```

```python
plt.plot(x,y,'g+')
```

**Add a format string as a third argument**

```python
plt.plot(x,y,'r+--')
```

```python
plt.plot(x,y,'--r+')
```

**Note one more thing the order does not matter in the format strings**

```python
plt.plot(x,y,'gD-.')
```

**You can do the same things with the properties**

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,color="red",linestyle='',marker="+")
```

```python
plt.plot(x,y,color="red",linestyle='',marker="D",markersize=15)
```

```python
plt.plot(x,y,color="#345678",linestyle='',marker="D",markersize=15)
```

**So you can control the color by giving hexadecimal number and control the rgb combinations , first 2 for R next 2 for G and last 2 for B**

**There is another property of alpha to control the transperancy of your plot ... Its value is between 0 and 1**

```python
plt.plot(x,y,color="#345678",marker=".",alpha=0.5)
```
