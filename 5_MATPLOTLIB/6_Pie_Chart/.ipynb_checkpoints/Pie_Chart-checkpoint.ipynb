# Pie Chart

Pie chart is usedwhen you want to show the individual thing / pie as a percentage of the total 

```python
import matplotlib.pyplot as plt
```

Basically the pyplot is the main object to draw all the charts 

```python
expense=[1200,5000,2000,6000,3000]
category=["internet","uni","petrol","bills","food"]
```

Basically for pie you need to pass the expense only and their label separately in the labels argument

```python
plt.pie(expense,labels=category)
```

To avoid the above documentation 

```python
plt.pie(expense,labels=category)
plt.show()
```

If some times the pie charts comes in the little bit oval shape do use the following axis()

```python
plt.axis("equal")
plt.pie(expense,labels=category)
plt.show()
```

Now to show the percentages inthe chart use autopct keyword , it takes the format string

```python
plt.pie(expense,labels=category,autopct='%f%%')  # before f tell what kind of format u need in float / decimal
plt.show()
```

```python
plt.pie(expense,labels=category,autopct='%0.2f%%')  # before f tell what kind of format u need in float / decimal
plt.show()
```

```python
plt.pie(expense,labels=category,autopct='%0.1f%%')  # before f tell what kind of format u need in float / decimal
plt.show()
```

If wanting to increase the size then use radius keyword

```python
plt.pie(expense,labels=category,autopct='%0.1f%%',radius=2)  # before f tell what kind of format u need in float / decimal
plt.show()
```

```python
plt.pie(expense,labels=category,autopct='%0.1f%%',radius=1.5)  # before f tell what kind of format u need in float / decimal
plt.show()
```

```python
# For fancy use shadow keyword
plt.pie(expense,labels=category,autopct='%0.1f%%',radius=1.2 , shadow=True)  # before f tell what kind of format u need in float / decimal
plt.show()
```

```python
# Now if want to make these pie outside 
plt.pie(expense,labels=category,autopct='%0.1f%%',radius=1.2,explode=[0,0,0,0,0]) # In exlode  give list of amount  of expenses , As we have 5 expense So ,
plt.show()
```

```python
# Now if want to make these pie outside 
plt.pie(expense,labels=category,autopct='%0.1f%%',radius=1.2,explode=[0.1,0,0,0,0.1]) # In exlode  give list of amount  of expenses , As we have 5 expense So ,
plt.show()
```

By default Pie Chart starts from first expense and at an angle of 0 and in counter clock wise direction
To change that use startangle argument

```python
plt.pie(expense,labels=category,autopct='%0.1f%%',radius=1.2,startangle=90) 
plt.show()
```
