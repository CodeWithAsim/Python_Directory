# Ploting Tutorial

```python
import matplotlib.pyplot as plt
import pandas as pd
```

**Creating two python lists**

```python
x=[1,2,3,4,5,6,7]
y=[34,35,46,27,45,44,38]
```

```python
%matplotlib inline
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

```python
plt.plot(x,y,color="green")
```

```python
from importlib import reload
plt=reload(plt)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='red')
```

```python
plt=reload(plt)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y,color="red",linestyle="dotted")
```

```python
plt.plot(x,y,color="red",linestyle="dashdot")
```

**For more about plot visit the plot api ... https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html**

# Ploting using DataFrame

```python
weather={
    "Days":[1,2,3,4,5,6,7],
    "Temperature":[34,35,46,27,45,44,38]
}
```

```python
df=pd.DataFrame(weather)
```

```python
df
```

```python
df.plot(y="Temperature")
```

```python
df.plot.bar(y="Temperature")
```
