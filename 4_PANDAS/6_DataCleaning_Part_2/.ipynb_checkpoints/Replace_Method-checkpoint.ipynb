# <center><h1>Replace Method for Performing Data Wrangling</h1></center>

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('weather_data.csv')
df
```

**Now we will clean this data using the replace method**

```python
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
df=pd.read_csv('weather_data.csv')
df
```

# <Font color="red">Using list of data to be replaced</font> 

```python
new_df=df.replace([-99999,-88888],np.NaN)
new_df
```

```python
new_df=df.replace([-99999,-88888,'0'],np.NaN)
new_df
```

# <Font color="red">Using dictionaries in Replace</font> 

```python
new_df=df.replace({
    "temperature":-99999,
    "windspeed":-88888,
    "event":'0'
},np.NaN)
new_df
```

# <Font color="red">Using Maping For Replacing</font> 

```python
df=pd.read_csv('weather_data.csv')
df
```

```python
new_df=df.replace({
    -99999:0,
    -88888:0,
    '0':"No Event"
})
new_df
```

# <Font color="red">Use of Regex to remove the measurement units</font> 

```python
df=pd.read_csv('weather_data.csv')
df
```

```python
new_df=df.replace('[A-za-z]','',regex=True)
new_df
```

**So now events are also the alphabets and they got null**

```python
new_df=df.replace({
    "temperature":'[A-za-z]',
    "windspeed":'[A-za-z]',
    "event":"0"
},'',regex=True)
new_df
```

# <Font color="red">Replacing the lists of values with the list of values</font> 

```python
score_card={
    "Grade":["A","B","C","D","E","F"],
    "Gpa":[4,3.5,3.0,2.5,2.0,1.0]
}
```

```python
import pandas as pd
df1=pd.DataFrame(score_card)
df1
```
